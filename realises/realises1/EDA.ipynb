Дропаем лишнее (колонки и не мск)

```python
df_msk = df.copy()
df_msk.drop(columns=['Unnamed: 0', 'Тип', 'Телефоны', 'Дополнительно', 'Название ЖК', 'Серия дома', 'Ссылка на объявление', 'Площадь комнат, м2'], inplace=True)
df_msk = df_msk[df['Адрес'].str.contains('Москва')]
df_msk
```

```python
#Достаем рубли из цены, плюс переводим все в рубли
df_msk['Цена, руб'] = [float(row.split(' ')[0]) if 'руб' in row else float(row.split(' ')[0]) / 80.65 if '$' in row else float(row.split(' ')[0]) / 92.91 for row in df_msk['Цена']]
df_msk.drop(columns='Цена', inplace=True)
```

```python
#достаем количесвто комнат
df_msk.loc[:, 'Количество комнат'] = [int(row.split(',')[0]) if type(row) == str else row for row in df_msk['Количество комнат']]

```

```python
# Убираем нан в метро, для обработки удаленности
df_msk.loc[:,'Метро'] =  df_msk['Метро'].fillna('Нету')
df_msk.loc[:,'Удаленность от метро, мин'] = [float(row[row.find('(') + 1:row.find(' мин'):]) if row[row.find('(') + 1:row.find(' мин'):].isdigit() else 0  for row in df_msk['Метро'] ]
```

```python
# достаем метро
df_msk['Метро'] = [row.split(' (', 1)[0] if row.split(' (', 1)[0] != 'Нету' else 'Нету'  for row in df_msk['Метро']]
df_msk['Метро'] = [row.split('. ', 1)[1] if '.' in row else 'Нету' for row in df_msk['Метро']]
df_msk['Метро'] = df_msk['Метро'].astype(str)
```

```python
# достаем Площадь общую
df_msk['Площадь, м2'] = [row.split('/')[0] for row in df_msk['Площадь, м2']]
df_msk['Площадь, м2'] = df_msk['Площадь, м2'].astype(float)
```

```python
# достаем этажность и дом
df_msk['Этажность'] = [row.split('/')[1] for row in df_msk['Дом']]
df_msk['Этажность'] = [row.split(',')[0] for row in df_msk['Этажность']]
df_msk['Дом'] = [row.split(',', 1)[1] if ',' in row else np.nan for row in df_msk['Дом']]
```

```python
df_msk['Ремонт'] = df_msk['Ремонт'].fillna('Без ремонта')
```

```python
def get_pass(row):
    if not isinstance(row, str):
        return 0
    if "Пасс" in row:
        return int(row.split("Пасс")[1].split("(")[1].split(")")[0])
    return 0

def get_gruz(row):
    if not isinstance(row, str):
        return 0
    if "Груз" in row:
        return int(row.split("Груз")[1].split("(")[1].split(")")[0])
    return 0

df_msk['Лифт пассажирский'] = df_msk['Лифт'].apply(get_pass)
df_msk['Лифт грузовой'] = df_msk['Лифт'].apply(get_gruz)
```

```python
def get_sovm(row):
    if not isinstance(row, str):
        return 0
    if "Совмещенный" in row:
        return int(row.split("Совмещенный")[1].split("(")[1].split(")")[0])
    return 0

def get_razd(row):
    if not isinstance(row, str):
        return 0
    if "Раздельный" in row:
        return int(row.split("Раздельный")[1].split("(")[1].split(")")[0])
    return 0

df_msk['Санузел совместный'] = df_msk['Санузел'].apply(get_sovm)
df_msk['Санузел раздельный'] = df_msk['Санузел'].apply(get_razd)
```

```python
df_msk['Можно с детьми'] = df_msk['Можно с детьми/животными'].apply(lambda x: 'Да' if isinstance(x, str) and 'детьми' in x else 'Нет')

# Разрешено с животными
df_msk['Можно с животными'] = df_msk['Можно с детьми/животными'].apply(lambda x: 'Да' if isinstance(x, str) and 'животными' in x else 'Нет')

df_msk.drop(columns=['Санузел', 'Можно с детьми/животными', 'Лифт'], inplace=True)

df_msk
```

```python
def plot_mean_price(column):
    plt.figure(figsize=(12, 6))

    grouped_mean = df_msk.groupby(column)['Цена, руб'].mean()

    plt.bar(grouped_mean.index, grouped_mean.values)
    plt.title(f'Влияние цены на {column}')
    plt.xlabel(column)
    plt.ylabel('Цена (руб.)')

    if column in ('Этажность', 'Дом'):
        plt.xticks(rotation=90)

    if column == 'Balcony':
        plt.title(f'Влияние цены на кол-во балконов')
    if column == 'Lodgia':
        plt.title(f'Влияние цены на кол-во лоджия')
    
    plt.grid(True)
    plt.tight_layout()

    plt.savefig(f'img/plot_{column}.png')
    plt.show()
    plt.close

category = [
    'Количество комнат', 'Площадь, м2', 'Дом', 'Парковка', 'Ремонт', 'Lodgia',
    'Balcony', 'Окна', 'Высота потолков, м', 'Мусоропровод', 'Удаленность от метро, мин',
    'Этажность', 'Лифт пассажирский', 'Лифт грузовой', 'Санузел совместный',
    'Санузел раздельный', 'Можно с детьми', 'Можно с животными', 'Округ'
]

for i in category:
    plot_mean_price(i)
```
